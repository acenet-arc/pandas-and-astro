## What's a "data frame"?

The data frame is a table with cases (or samples, or instances) in the rows.  Each case has one or more observations (or measurements, or features, or fields) in the columns.  The measurements may be numerical, or categorical, or textual, or even missing.  This is one of the things that distinguishes a data frame from a NumPy (or C or Fortran) array, in which all the values must be of a single type. 

Another thing that distinguishes a data frame (at least in pandas) is that the rows can be labelled, rather than just using integer indices. 

*  Would the first column (`galaxy`) be appropriate labels for these data, or not? Is there a field that would make better labels?

There's nothing (except chance) preventing two supernovas in one galaxy, but case labels should be unique. While we have no duplicate `galaxy` entries right here right now, it still seems like a poor choice in the long run.  On the other hand the supernova designations *are* meant to be universally unique, so let's use `supernova` as the case label. (Second column, which is column index 1, because zero-based indexing.)  This'll be the last time we re-load this data file, I promise :-)

```python
data = pandas.read_csv('hubble_data.dat', delim_whitespace=True, index_col=1,
                       names=['galaxy', 'supernova', 'm', 'sig_m',  'dist_mod', 'sig_dist_mod', 'M', 'sig_M', 'velocity'])
data
```

## Visualization

Pandas uses `matplotlib`, the same plotting package used in the rest of the Numpy/Scipy community.

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(data['velocity'], data['dist_mod'], 'o')
```

There's the plot we want, sort of.  Problem is that the x-axis is not distance but distance modulus (https://lco.global/spacebook/what-is-distance-modulus/).  What we'd like to plot is velocity in km/sec versus distance in Mpc.  Here's where you can put your Python training to use:
* Write a function which converts distance_modulus into distance.

```python
def distance_from_modulus(dist_mod):
    parsecs = 10**((dist_mod+5)/5)
    return parsecs
```

```python
data['dist_mod']
```

```python
distance = distance_from_modulus(data['dist_mod'])
```

```python
print(distance)
```

Those were in ParSecs, convert them to MegaParSecs

```python
print(distance/1.e6)
```

We can add the distances to the data frame easily:

```python
data['distance'] = distance/1.e6
```

```python
data
```

```python
plt.plot(data['velocity'], data['distance'], 'o')
```

Now we want to find a straight-line fit for this.  How do we find a function that does this?  Google throws up several suggestions, including numpy.polyfit, something from scipy, something else from scikit-learn, and statsmodels.  Based on the advice of basically the first web site I looked at, let's try statsmodels:

```python
import statsmodels.api as sm
```

```python
x = data['distance']
y = data['velocity']
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
```

```python
print(model.predict(x))
```

```python
model.summary()
```

The output from `statsmodels.api` is a bit dense, but basically we're looking for the distance coef, 53.6297.  Problem here is we don't know what the units are on the `velocity` column of our dataset, but if it's in km/sec then the number is in line with current estimates of 50-80 km/s/Mpc.
